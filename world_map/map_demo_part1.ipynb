```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
# This code from 
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson())

# make the map global rather than have it zoom in to
# the extents of any plotted data
ax.set_global()
ax.stock_img() # stock_img is a global topographic/bathymetric raster
ax.coastlines()

ax.plot(-0.08, 51.53, 'o', transform=ccrs.PlateCarree())
ax.plot([-0.08, 132], [51.53, 43.17], transform=ccrs.PlateCarree())
ax.plot([-0.08, 132], [51.53, 43.17], transform=ccrs.Geodetic())

plt.show()
```

Next, let's focus on North America and make a figure illustrating a few different map projections.

You can see a complete list of the available map projections in cartopy here:

https://scitools.org.uk/cartopy/docs/latest/crs/projections.html

```python
# define latitude and longitude bounds for the plot
```

```python
plot_extent = [-170,-30,10,75] # [min_lon,max_lon,min_lat,max_lat]
# map projection - use a mercator projection
proj = ccrs.Mercator(min_latitude=plot_extent[2], max_latitude=plot_extent[3])

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,2,1,projection=proj)
ax.stock_img()
ax.coastlines()
ax.plot(-121.754526,38.540945, 'rx', transform=ccrs.PlateCarree(),label='Davis')
ax.set_extent(plot_extent, crs=ccrs.PlateCarree())
plt.legend()

# add a second subplot with a different projection, for comparison
proj = ccrs.LambertCylindrical()
ax = fig.add_subplot(1,2,2,projection=proj)
ax.stock_img()
ax.coastlines()
ax.plot(-121.754526,38.540945, 'rx', transform=ccrs.PlateCarree(),label='Davis')
ax.set_extent(plot_extent, crs=ccrs.PlateCarree())
plt.legend()

plt.show()
```
